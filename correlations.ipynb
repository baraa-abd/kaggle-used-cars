```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

training = pd.read_csv("cars_train_enriched_denoised.csv")
```

```
training=training.apply(lambda x : pd.factorize(x)[0] if x.dtype=='object'else x)
plt.figure(figsize=(22,1))
sns.heatmap(training.corr()[12:13],annot=True,cmap='coolwarm')
plt.show()
```

```
training=training.apply(lambda x : pd.factorize(x)[0] if x.dtype=='object'else x)
plt.figure(figsize=(22,1))
sns.heatmap(training.corrwith(training['price'].astype('float'),method=stats.pointbiserialr),annot=True,cmap='coolwarm')
plt.show()
```
